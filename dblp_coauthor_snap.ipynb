**Description**

A coauthorship network extracted from the DBLP computer science manuscript database, in 2012. This network is a one-mode projection from the bipartite graph of computer scientists and their publications.

```python
import networkx as nx
```

```python
# Load the GraphML file
G = nx.read_graphml('coauthor-dblp/network.graphml')
```

```python
type(G)
```

```python
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
print(f"Is directed? {G.is_directed()}")
```

```python
# View nodes and edges
print("Nodes:", list(G.nodes(data=True))[:5])  # first 5 nodes with attributes
print("Edges:", list(G.edges(data=True))[:5])  # first 5 edges with attributes
```

```python
# Accessing a specific node's attributes
for node, attr in G.nodes(data=True):
    print(f"Node {node} has attributes {attr}")
    break  # remove break to see all

# Accessing edge attributes
for u, v, attr in G.edges(data=True):
    print(f"Edge from {u} to {v} with attributes {attr}")
    break
```

```python
import matplotlib.pyplot as plt

# Extract a subgraph with the first 100 nodes
subgraph_nodes = list(G.nodes())[:100]
H = G.subgraph(subgraph_nodes)

# Draw the subgraph
plt.figure(figsize=(12, 12))
nx.draw(H, with_labels=True, node_size=50, font_size=8)
plt.show()
```

```python
import numpy as np

degrees = [val for (node, val) in G.degree()]
print(f"Average degree: {np.mean(degrees)}")
print(f"Max degree: {np.max(degrees)}")
```

```python
components = nx.connected_components(G)
largest_component = max(components, key=len)
print(f"Size of largest connected component: {len(largest_component)}")
```
